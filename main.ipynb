```python
# dataframe mgmt
import pandas as pd

# visualization
import streamlit as st
import plotly.express as px
import codecs
import streamlit.components.v1 as components

# import database
df = pd.read_csv('output/linkedin_skill_requirements_full_2022-09-01.csv')
df.sample()
```

```python
# visualize skill frequency per job
round(df.groupby('job').mean()*100, 2)
```

```python
# too many skills to see all. Try checking by most popular skills
job_to_filter = input() # use 'Data Scientist' for example
num_jobs = len(df.loc[df['job'] == job_to_filter])

df_filtered = df.loc[df['job'] == f'{job_to_filter}'].groupby('job').sum().T
print(f'Total {job_to_filter} jobs: {num_jobs}')
df_top30 = df_filtered.sort_values(by=[f'{job_to_filter}'], ascending=False).head(30)
df_top30
```

```python
#data = px.data.df_top30
fig = px.bar(df_top30, x='job', y='Data Scientist')
fig.show()
```
